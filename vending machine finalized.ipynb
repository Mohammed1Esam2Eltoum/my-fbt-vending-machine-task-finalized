```python
products = {
    "snicker": 2, "water": 1, "fanta": 3, "lays": 5,
    "holsten": 6.5, "code red": 5, "bubble gum": 1,
    "potato chips": 7, "barbicana": 8, "cat food": 15,
    "dog food": 15, "tissues": 0.5, "tissue fine": 1.5
}


while True:
    #Shows the Menu
    print("\nAvailable Products:")
    for name, price in products.items():
        print(f"- {name.capitalize()}: {price} SR")
    
    
    item = input("\nEnter product name (or type 'exit' to quit): ").lower() #asks you the item you want
    
    if item == 'exit':
        print("Goodbye")
        break #to stop the infinite loop created by while true
        
    if item in products:
        price = products[item]
        print(f"Selected: {item.capitalize()} | Price: {price} SR")
        
        
        try: #so that if the code crashes or the purchase gets stopped it jumps to the except command
            cash = float(input("Insert cash (SR): "))
            
            if cash >= price:
                change = cash - price
                print("--- Processing ---")
                print(f"Dispensing {item}... Done")
                print(f"purchase complete. Your change is: {change:.2f} SR") #used .2f so that the code brings me the change up to the 2nd decimal
            else:
                print(f"Error: Insufficient funds. You need {price - cash:.2f} SR more.")
                
        except ValueError: #ties back to the try command incase the code fails
            print("Invalid input! Please insert real money (numbers only).")
    else:  #makes it so that the code doesn't crash if the user inputs an item not in the dictionary 
        print("Invalid selection. Please pick something from the list.")

    print("-" * 30)
```
